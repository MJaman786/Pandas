## iloc Use

In Pandas, .iloc is a powerful indexer used for selecting rows and columns by integer-based indexing. 
Unlike .loc, which is label-based (it selects data based on explicit labels in the index or column headers), 
.iloc uses integer positions to identify data.

Here's a detailed look at how .iloc works:

![](Datasets/sec1.png)

![](Datasets/sec22.png)

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 22],
    'Salary': [50000, 60000, 70000, 80000, 45000]
}
df = pd.DataFrame(data)
df
```

```python
# Select the first row (index 0)
print(df.iloc[0])
```

![](Datasets/sec2.1.png)

```python
# Select the first row (index 0)
print(f"{df.iloc[0]}\n")

# Select rows at positions 1, 3, and 4
print(f"{df.iloc[[1, 3, 4]]}\n")

# Select the first three rows
print(f"{df.iloc[:3]}\n")

"""----"""
# Select the first row and second column (Age)
print(f"{df.iloc[0, 1]}\n")  # Output: 25

# Select rows at index positions 1 and 2, and columns at positions 0 and 2
print(df.iloc[[1, 2], [0, 2]])
```

![](Datasets/sec3.png)

```python
# Select rows from index 1 to 3 and columns from index 0 to 2
print(f"{df.iloc[1:4, 0:2]}\n")

# Select all rows and first column
print(f"{df.iloc[:, 0]}\n")  # Outputs the 'Name' column

# Select all columns for row 2
print(f"{df.iloc[2, :]}\n")  # Outputs all columns for 'Charlie'
```
